```python
dataset = pd.read_excel("teste_revisores.xlsx",sheet_name = "Planilha8") #Regressão,ensemble,pycaret
#dataset = litoral.drop('Município',axis=1)
array = dataset.values
X = array[:, 4:].astype(float)
Y = array[:, 3]
Y
test_size = 0.30
seed = 7

pipe = Pipeline([('preprocessing', StandardScaler()), ('classifier', MLPClassifier(max_iter=1000, verbose = True, tol = 0.0001, alpha=1, hidden_layer_sizes = (100,100)))])
param_grid = [
    {'classifier': [ MLPClassifier(max_iter=1000, verbose = True, tol = 0.0001, alpha=1, hidden_layer_sizes = (100,100))], 
     'preprocessing': [StandardScaler()]},
    {'classifier':[RandomForestClassifier(bootstrap=False, ccp_alpha=0.0, #criterion='mse',
                      max_depth=2, max_features=10, max_leaf_nodes=None, # max_features etava em 'auto'
                      max_samples=None, min_impurity_decrease=0.0,
                      min_samples_leaf=1,
                      min_samples_split=2, min_weight_fraction_leaf=0.0,
                      n_estimators=1000, n_jobs=-1, oob_score=False,
                      random_state=42, verbose=0, warm_start=False)], 
     'preprocessing': [None]}]
#,None],
 # 'classifier__gamma': [0.001, 0.01, 0.1, 1, 10, 100],
 # 'classifier__C': [0.001, 0.01, 0.1, 1, 10, 100]},
 # {'classifier': [RandomForestClassifier(n_estimators=100)],
 # 'preprocessing': [None], 'classifier__max_features': [1, 2, 3]}]

    # ('Scaler', StandardScaler()),
    # ('MLP', MLPClassifier(max_iter=1000, verbose = True, tol = 0.0001, alpha=1, hidden_layer_sizes = (100,100)))])))

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=test_size, random_state=seed)

grid = GridSearchCV(pipe, param_grid, cv=4)
grid.fit(X_train, y_train)
print("Best params:\n{}\n".format(grid.best_params_))
print("Best cross-validation score: {:.2f}".format(grid.best_score_))
print("Test-set score: {:.2f}".format(grid.score(X_test, y_test)))
```
